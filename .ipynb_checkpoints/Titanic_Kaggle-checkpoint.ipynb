```python
del test_df['Cabin']
del train_df['Cabin']
```

```python
all_df = train_df.append(test_df)
```

```python
# 빈 값이 많은 Cabin을 없앤 후, NaN값이 많은 값을 확인
(all_df.isnull().sum() / len(all_df)).plot(kind='bar')
plt.show()
```

```python
len(all_df)
```

```python
# 생존도와 크게 상관이 없는 티켓과 이름 삭제
del all_df["Name"]
del all_df["Ticket"]
```

```python
all_df.head()
```

```python
all_df["Sex"] = all_df["Sex"].replace({"male":0,"female":1})
all_df.head()
```

```python
#이거 왜 오류가 날까?
all_df['Sex'] = all_df['Sex'].replace({"male" : 0, "female" : 1})
```

```python
all_df
```

```python
all_df['Embarked'].unique()
```

```python
all_df['Embarked'] = all_df['Embarked'].replace({'S' : 0, 'C' : 1, 'Q' : 2, np.nan : 99})
```

```python
all_df
```

```python
pd.get_dummies(all_df["Embarked"], prefix="embarked")
```

```python
mat_df = pd.merge(
    all_df, pd.get_dummies(all_df["Embarked"], prefix = "embarked"),
    left_index=True, right_index=True)
mat_df.head()
```

```python
#correlation - 상관계수
mat_df.corr()
```

```python
# 각 클래스 -123 별 나이의 평균 : 클래스가 높을수록 나이가 많다 -> 돈이 많을 확률이 높다
all_df.groupby("Pclass")["Age"].mean()
```

```python
all_df.loc[
    (all_df['Pclass'] == 1) & (all_df["Age"].isnull()), "Age"]
```

```python
all_df.loc[
    (all_df['Pclass'] == 1) & (all_df["Age"] == 1), "Age"]
```

```python
all_df['Age'] = all_df['Pclass']
all_df["Pclass"]
```

```python
all_df["Age"]
```

```python
all_df.loc[(all_df["Pclass"] == 1), "Age"] = 39.16 
all_df.loc[(all_df["Pclass"] == 2), "Age"] = 29.51 
all_df.loc[(all_df["Pclass"] == 3), "Age"] = 24.82
```

```python
all_df
```

```python
all_df.isnull().sum()
```

```python
all_df.groupby("Pclass")["Fare"].mean()
```

```python
all_df[all_df["Fare"].isnull()]
```

```python
all_df.loc[all_df["Fare"].isnull(), "Fare"] = 13.30
```

```python
del all_df["Embarked"]
```

```python
all_df["Pclass"] = all_df["Pclass"].replace({1: "A", 2:"B", 3:"C"})
```

```python
all_df = pd.get_dummies(all_df)
all_df.head()
```

```python
#오류가 있음 확인!
all_df = pd.merge(
    all_df, mat_df[["embarked_0","embarked_1","embarked_2", "embarked_99"]],
    left_index=True, right_index = True)
all_df
```

```python
train_df = all_df[all_df.index.isin(train_index)]
test_df = all_df[all_df.index.isin(test_index)]
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
x_data = train_df.values
y_data = y_train_df.values
```

```python
x_data.shape, y_data.shape
```

```python
y_data
```

```python
from sklearn.linear_model import LogisticRegression
cls = LogisticRegression()
cls.fit(x_data,y_data)
```

```python
cls.intercept_
```

```python
cls.coef_
```

```python
cls.predict(test_df.values)
```

```python
test_df.index
```

```python
x_test = test_df.values
y_test = cls.predict(x_test)
y_test
```

```python
result = np.concatenate( (test_index.values.reshape(-1,1),
                            cls.predict(x_test).reshape(-1,1) ), axis =1)
result[:5]
```

```python
df_submission =pd.DataFrame(result, columns=["PassengerId", "Survived"])
df_submission
```

```python
df_submission.to_csv("submission_result.csv", index=False)
```
